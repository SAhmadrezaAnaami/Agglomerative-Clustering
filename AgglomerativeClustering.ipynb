```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = make_blobs(n_samples=100, centers=3,random_state=42)
x.shape , y.shape
```

```python
plt.scatter(x[:,0] , x[:,1])
plt.show()
```

```python
agg_clustering = AgglomerativeClustering(n_clusters=3)
agg_clustering.fit(x)
```

```python
# Predict the cluster labels for each data point
agg_clustering_labels = agg_clustering.labels_
print(agg_clustering_labels)
```

```python
plt.scatter(x[:,0] , x[:,1] , c = agg_clustering_labels , cmap='rainbow')
plt.show()
```

```python
def generate_data(m):
  """plots m random points on a 3D plane"""
  angles = np.random.rand(m) * 5 * np.pi / 2 - 0.5
  data = np.empty((m, 3))
  data[:,0] = np.cos(angles) + np.sin(angles)/2 + 0.1 * np.random.rand(m)/2
  data[:,1] = np.sin(angles) * 0.7 + 0.1 * np.random.rand(m) / 2

  return data
```

```python
data = generate_data(100)

plt.scatter(data[:,0] , data[:,1])
plt.show()
```

```python
agg2_clustering = AgglomerativeClustering(n_clusters=3)
agg2_clustering.fit(data)
```

```python
# Predict the cluster labels for each data point
agg2_clustering_labels = agg2_clustering.labels_
print(agg2_clustering_labels)
```

```python
plt.scatter(data[:,0] , data[:,1] , c = agg2_clustering_labels , cmap="rainbow")
plt.show()
```

```python
def generate_data_3d(m):
  """plots m random points on a 3D plane"""
  angles = np.random.rand(m) * 5 * np.pi / 2 - 0.5
  data = np.empty((m, 3))
  data[:,0] = np.cos(angles) + np.sin(angles)/2 + 0.1 * np.random.rand(m)/2
  data[:,1] = np.sin(angles) * 0.7 + 0.1 * np.random.rand(m) / 2
  data[:,2] = data[:,0] * 0.1 + data[:,1] * 0.3 + 0.1 * np.random.rand(m)

  return data
```

```python
# use the function above to generate data points
X_train = generate_data_3d(200)


# preview the data
ax = plt.axes(projection="3d")
ax.scatter3D(X_train[:, 0], X_train[:, 1], X_train[:, 2])
```

```python
agg2_clustering_3d = AgglomerativeClustering(n_clusters=4)
agg2_clustering_3d.fit(X_train)
```

```python
# Predict the cluster labels for each data point
agg2_clustering_3d_labels = agg2_clustering_3d.labels_
print(agg2_clustering_3d_labels)
```

```python
# preview the data
ax = plt.axes(projection="3d")
ax.scatter3D(X_train[:, 0], X_train[:, 1], X_train[:, 2] , c = agg2_clustering_3d_labels ,cmap = "rainbow")
```
